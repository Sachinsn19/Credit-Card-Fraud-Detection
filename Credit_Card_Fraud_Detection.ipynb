# Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import  train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

# Data Collection and Analysis

Dataset : [credit card.csv](https://www.kaggle.com/mlg-ulb/creditcardfraud)

```python
#Loading dataset into Pandas Dataframe
credit_card_data = pd.read_csv("/content/creditcard.csv")
```

```python
#Printing top 5 rows of the data
credit_card_data.head()
```

```python
#Printing last 5 rows of the data
credit_card_data.tail()
```

```python
#Dataset informations
credit_card_data.info()
```

```python
#Checking the number of missing values
credit_card_data.isnull().sum()
```

```python
#Distribution of legit transformation and fraudulent transformations
credit_card_data['Class'].value_counts()
```

The data is highly unbalanced

0 --> Legit Transformation

1 --> Fradualent Transformation

```python
#Seperating the data for analysis
legit  = credit_card_data[credit_card_data['Class']==0]
fraud = credit_card_data[credit_card_data['Class']==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the legit data
legit.Amount.describe()
```

```python
#statistical measures of the fraud data
fraud.Amount.describe()
```

```python
#Compare values for  both transactions
credit_card_data.groupby('Class').mean()
```

## Under-Sampling

Build a sample dataset containing  similar distribution of legit and fraudulent transactions

Fraudulant transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

```python
#Concatenating  two dataframes
new_dataset = pd.concat([legit_sample,fraud],axis=0) 
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.Class.value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
credit_card_data.groupby('Class').mean()
```

### Splitting the data into Features and Target

```python
X = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
print(X)
print(y)
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, stratify=y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

# Model Evaluation

### Accuracy score

```python
#Accuracy score on training data
train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(y_train, train_prediction)
```

```python
print("Accuracy score on Training data : ",train_data_accuracy)
```

```python
#Accuracy score on training data
test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(y_test, test_prediction)
```

```python
print("Accuracy score on Test data : ",test_data_accuracy)
```
